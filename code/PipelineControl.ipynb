##Transformation Control

###Base Tranformation Control

```python
class TransformationControl:
    # Transform Metrices
    TRANSFORM_METRIX_PULSE_CHECK:str = 'source-pulse-check';
    TRANSFORM_METRIX_EXTRACT_CHECK:str = 'extraction-check';
    TRANSFORM_METRIX_RELATION_CHECK:str = 'relationship-check';
    TRANSFORM_METRIX_TRANSFORM_STAT:str = 'transformation-status';
    TRANSFORM_METRIX_LOAD_STAT:str = 'load-status';

    RELATION_COLUMNS:list = ['tgt_entity', 'ref_entity', 'tgt_fk', 'ref_pk'];

    mandatory_entities:list = None;
    optional_entitie:list = None;

    mandatory_relations:pd.DataFrame = None;
    optional_relations:pd.DataFrame = None;

    # Overall metrices
    metrix:pd.DataFrame = None;

    # Process metrices
    data_retrieval_metrix:pd.DataFrame = None;
    relation_metrix:pd.DataFrame = None;
    transformed_dataset:pd.DataFrame = None;

    transformed_entity_label:str = None

    def init_metrix(self):
        ls:list = [self.TRANSFORM_METRIX_PULSE_CHECK, self.TRANSFORM_METRIX_EXTRACT_CHECK,
                self.TRANSFORM_METRIX_RELATION_CHECK, self.TRANSFORM_METRIX_TRANSFORM_STAT,
                self.TRANSFORM_METRIX_LOAD_STAT];
        ls_len = len(ls);
        self.metrix = pd.DataFrame({
                'metrice': ls,
                'mandatory': [None]*ls_len,
                'verified_mandatory': [None]*ls_len,
                'optional': [None]*ls_len,
                'verified_optional': [None]*ls_len,
                'df_mandatory': [None]*ls_len,
                'df_optional': [None]*ls_len});

    def get_data_retrieval_report(self) -> pd.DataFrame: return self.data_retrieval_metrix.loc[:];
    def get_relation_metrix_report(self) -> pd.DataFrame: return self.relation_metrix.loc[:];

    def retrieve_data(self) -> pd.DataFrame:
        def extract(df:pd.DataFrame)->pd.DataFrame:
            df = df.loc[:];
            if df['pulse'] == 1:
                try:
                    with df['blob'].open() as csv:
                        df['df'] = pd.read_csv(csv);
                        df['extract'] = 1;
                except Exception as e:
                    df['err_flag'] = 1;
                    df['err_desc'] = str(e);
            return df.loc[:];

        retrieval_metrix:pd.DataFrame = pd.DataFrame({
            'entity': self.mandatory_entities + self.optional_entities,
            'ty': ['mandatory']*len(self.mandatory_entities) + ['optional']*len(self.optional_entities),
        });
        retrieval_metrix['pulse'] = 0;
        retrieval_metrix['extract'] = 0;
        retrieval_metrix['err_flag'] = 0;
        retrieval_metrix['err_desc'] = None;
        retrieval_metrix['df'] = None;

        stored = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_STORAGE);

        # Check pulse
        retrieval_metrix['blob'] = retrieval_metrix['entity'].apply(lambda x: stored.blob(x+'.csv'));
        retrieval_metrix['pulse'] = retrieval_metrix['blob'].apply(lambda x: 1 if x.exists() else 0);

        # Extract data
        retrieval_metrix = retrieval_metrix.apply(extract, axis='columns');
        retrieval_metrix.drop(columns=['blob'], inplace=True);

        self.data_retrieval_metrix = retrieval_metrix.loc[:];

        # Report metrices
        mandated_metrix = retrieval_metrix[retrieval_metrix['ty']=='mandatory'].loc[:];
        optional_metrix = retrieval_metrix[retrieval_metrix['ty']=='optional'].loc[:];
        self.metrix.set_index('metrice', inplace=True);
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'mandatory'] = len(mandated_metrix.index);
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'verified_mandatory'] = mandated_metrix['pulse'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'optional'] = len(optional_metrix.index);
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'verified_optional'] = optional_metrix['pulse'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'df_mandatory'] = mandated_metrix;
        self.metrix.at[self.TRANSFORM_METRIX_PULSE_CHECK, 'df_optional'] = optional_metrix;

        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'mandatory'] = len(mandated_metrix.index);
        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'verified_mandatory'] = mandated_metrix['extract'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'optional'] = len(optional_metrix.index);
        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'verified_optional'] = optional_metrix['extract'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'df_mandatory'] = mandated_metrix;
        self.metrix.at[self.TRANSFORM_METRIX_EXTRACT_CHECK, 'df_optional'] = optional_metrix;
        self.metrix.reset_index(names=['metrice'], inplace=True);

        return self.metrix.loc[:];

    def check_relation(self):
        def verify_relation(relation:pd.DataFrame) -> int:
            relation = relation.loc[:];
            if (relation['tgt_ext'] == 0) or (relation['ref_ext'] == 0): return 0;
            tgt_df:pd.DataFrame = relation['tgt_df'];
            ref_df:pd.DataFrame = relation['ref_df'];

            tgt_fk:pd.Series = tgt_df[relation['tgt_fk']].drop_duplicates();
            ref_pk:pd.Series = ref_df[relation['ref_pk']].drop_duplicates();

            valid_fk:pd.DataFrame = tgt_fk[tgt_fk.isin(ref_pk)];
            return 1 if len(tgt_fk)==len(valid_fk) else 0;

        def reason_rejection(relation:pd.DataFrame) -> str:
            relation = relation.loc[:];
            if (relation['tgt_ext'] == 0) and (relation['ref_ext'] == 0):
                return '{} and {} are not available.'.format(relation['tgt_entity'], relation['ref_entity']);
            if (relation['tgt_ext'] == 0): return '{} is not available.'.format(relation['tgt_entity']);
            if (relation['ref_ext'] == 0): return '{} is not available.'.format(relation['ref_entity']);
            if (relation['verified'] ==0):
                return '{} - {} relation mismatched '.format(relation['tgt_entity'], relation['ref_entity']);

        retrieved = self.data_retrieval_metrix.loc[:]

        man_rel:pd.DataFrame = self.mandatory_relations.loc[:];
        man_rel['ty'] = 'mandatory';
        man_rel['verified'] = 0;
        man_rel['reason'] = None;

        opt_rel:pd.DataFrame = self.optional_relations.loc[:];
        opt_rel['ty'] = 'optional';
        opt_rel['verified'] = 0;
        opt_rel['reason'] = None;

        # Verify relation between mandatory entity
        man_rel = (man_rel.set_index('ref_entity')
                .join(retrieved[['entity', 'df', 'extract']].set_index('entity'))
                .reset_index(names=['ref_entity'])
                .rename(columns={'df':'ref_df', 'extract':'ref_ext'}));
        man_rel = (man_rel.set_index('tgt_entity')
                .join(retrieved[['entity', 'df', 'extract']].set_index('entity'))
                .reset_index(names=['tgt_entity'])
                .rename(columns={'df':'tgt_df', 'extract':'tgt_ext'}));
        man_rel['verified'] = man_rel.apply(verify_relation, axis='columns');
        man_rel['reason'] = man_rel.apply(reason_rejection, axis='columns');
        man_rel.drop(columns=['tgt_df', 'tgt_ext', 'ref_df', 'ref_ext'], inplace=True);

        # Verify relation related to optional entity
        if len(opt_rel.index) > 0:
            opt_rel = (opt_rel.set_index('ref_entity')
                    .join(retrieved[['entity', 'df', 'extract']].set_index('entity'))
                    .reset_index(names=['ref_entity'])
                    .rename(columns={'df':'ref_df', 'extract':'ref_ext'}));
            opt_rel = (opt_rel.set_index('tgt_entity')
                    .join(retrieved[['entity', 'df', 'extract']].set_index('entity'))
                    .reset_index(names=['tgt_entity'])
                    .rename(columns={'df':'tgt_df', 'extract':'tgt_ext'}));
            opt_rel['verified'] = opt_rel.apply(verify_relation, axis='columns');
            opt_rel['reason'] = opt_rel.apply(reason_rejection, axis='columns');
            opt_rel.drop(columns=['tgt_df', 'tgt_ext', 'ref_df', 'ref_ext'], inplace=True);

        self.relation_metrix = pd.concat([man_rel, opt_rel]);

        # Report metrices
        self.metrix.set_index('metrice', inplace=True);
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'mandatory'] = len(man_rel.index);
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'verified_mandatory'] = man_rel['verified'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'optional'] = len(opt_rel.index);
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'verified_optional'] = opt_rel['verified'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'df_mandatory'] = man_rel;
        self.metrix.at[self.TRANSFORM_METRIX_RELATION_CHECK, 'df_optional'] = opt_rel;
        self.metrix.reset_index(names=['metrice'], inplace=True);

        return self.metrix.loc[:]

    def transform(self) -> pd.DataFrame:
        prev_proc_ls = [self.TRANSFORM_METRIX_PULSE_CHECK, self.TRANSFORM_METRIX_EXTRACT_CHECK, self.TRANSFORM_METRIX_RELATION_CHECK];
        ctrl_metrix = self.metrix[['metrice', 'mandatory', 'verified_mandatory']].loc[:];
        ctrl_metrix = ctrl_metrix[ctrl_metrix['metrice'].isin(prev_proc_ls)].loc[:]
        ctrl_metrix['pass_flag'] = ( ctrl_metrix[['mandatory', 'verified_mandatory']]
                .apply(lambda x: 1 if (x['mandatory']==x['verified_mandatory']) else 0, axis='columns') );

        df:pd.DataFrame = None;
        if ( (ctrl_metrix['pass_flag'].sum() == len(ctrl_metrix.index))
                and (not ctrl_metrix['mandatory'].isnull().values.any())
                and (not ctrl_metrix['verified_mandatory'].isnull().values.any()) ):
            df = self.transform_core();

        self.transformed_dataset = df.loc[:] if df is not None else None;

        # Report metrice
        self.metrix.set_index('metrice', inplace=True);
        self.metrix.at[self.TRANSFORM_METRIX_TRANSFORM_STAT, 'mandatory'] = 1;
        self.metrix.at[self.TRANSFORM_METRIX_TRANSFORM_STAT, 'verified_mandatory'] = 1 if df is not None else 0;
        self.metrix.at[self.TRANSFORM_METRIX_TRANSFORM_STAT, 'df_mandatory'] = df;
        self.metrix.reset_index(names=['metrice'], inplace=True);

        return self.metrix.loc[:];

    # Callable function to overrid
    def transform_core(self) -> pd.DataFrame: return None;


    def dump_data(self, prev_proc:list=None) -> pd.DataFrame:
        prev_proc_ls = ( prev_proc if prev_proc is not None else
                [self.TRANSFORM_METRIX_PULSE_CHECK, self.TRANSFORM_METRIX_EXTRACT_CHECK,
                self.TRANSFORM_METRIX_RELATION_CHECK, self.TRANSFORM_METRIX_TRANSFORM_STAT] );

        ctrl_metrix = self.metrix[['metrice', 'mandatory', 'verified_mandatory']].loc[:];
        ctrl_metrix = ctrl_metrix[ctrl_metrix['metrice'].isin(prev_proc_ls)].loc[:]
        ctrl_metrix['pass_flag'] = ( ctrl_metrix[['mandatory', 'verified_mandatory']]
                .apply(lambda x: 1 if (x['mandatory']==x['verified_mandatory']) else 0, axis='columns') );

        df:pd.DataFrame = pd.DataFrame(columns=['entity', 'verified', 'err_flag', 'reason'],
                data=[[self.transformed_entity_label, 0, 0, None]]);

        if ( (ctrl_metrix['pass_flag'].sum() == len(ctrl_metrix.index))
                and (not ctrl_metrix['mandatory'].isnull().values.any())
                and (not ctrl_metrix['verified_mandatory'].isnull().values.any()) ):

            stored = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_STORAGE);
            blob = stored.blob(self.transformed_entity_label+'.csv');

            df.set_index('entity', inplace=True);
            try:
                blob.upload_from_string(self.transformed_dataset.to_csv(index=False));
                df.at[self.transformed_entity_label, 'verified'] = 1;
            except Exception as e:
                df.at[self.transformed_entity_label, 'err_flag'] = 1;
                df.at[self.transformed_entity_label, 'reason'] = str(e);
            df.reset_index(names=['entity'], inplace=True);

        # Report metrice
        self.metrix.set_index('metrice', inplace=True);
        self.metrix.at[self.TRANSFORM_METRIX_LOAD_STAT, 'mandatory'] = len(df.index);
        self.metrix.at[self.TRANSFORM_METRIX_LOAD_STAT, 'verified_mandatory'] = df['verified'].sum();
        self.metrix.at[self.TRANSFORM_METRIX_LOAD_STAT, 'df_mandatory'] = df;
        self.metrix.reset_index(names=['metrice'], inplace=True);

        return self.metrix.loc[:];

    def execute(self):
        self.retrieve_data();
        self.check_relation();
        self.transform();
        self.dump_data();
        return self.metrix.loc[:];
```

###ML Customer Classification Control Based-class

```python
class MLCustomerClassificationControl(PipelineControl):
    classification:mlmodel.CustomerClassification = None;
    k_mean__k:int = 4;

    def set_k_mean__k(self, k:int):
        self.k_mean__k = k;
        return self;


    def build_model(self):
        def extraction_src(entity:storage.Blob, metrix:pd.DataFrame):
            if not entity.exists():
                metrix.at['src-extract', 'err_flag'] = 1;
                metrix.at['src-extract', 'err_desc'] = 'Blob is not existed';
                return None, metrix.loc[:];
            try:
                with entity.open() as csv:
                    metrix.at['src-extract', 'progress'] = 1;
                    return pd.read_csv(csv), metrix;
            except Exception as e:
                metrix.at['src-extract', 'err_flag'] = 1;
                metrix.at['src-extract', 'err_desc'] = str(e);
            return None, metrix.loc[:];

        def build(metrix:pd.DataFrame, df:pd.DataFrame)->pd.DataFrame:
            if (metrix['err_flag'].sum() > 0) or (df is None): return metrix.loc[:];
            try:
                ( mlmodel.CustomerClassification().set_profile(df).prepare_trainset()
                        .set_model_k(self.k_mean__k).train_model()
                        .export_model(self.ML_MODEL_CUST_CLASS+'.joblib') );
                metrix.at['build-model', 'progress'] = 1;
            except Exception as e:
                metrix.at['build-model', 'err_flag'] = 1;
                metrix.at['build-model', 'err_desc'] = str(e);
            return metrix.loc[:];

        def store(metrix:pd.DataFrame)->pd.DataFrame:
            if metrix['err_flag'].sum() > 0: return metrix.loc[:];

            try:
                obj_name:str = self.ML_MODEL_CUST_CLASS+'.joblib';
                ( storage.Client(project=self.GCP_PROJECT_ID)
                        .bucket(self.GCP_STORAGE_BUCKET_ML_MODEL)
                        .blob(obj_name)
                        .upload_from_filename(obj_name) );
                metrix.at['store-model', 'progress'] = 1;
            except Exception as e:
                metrix.at['store-model', 'err_flag'] = 1;
                metrix.at['store-model', 'err_desc'] = str(e);
            return metrix.loc[:];

        metrix:pd.DataFrame = pd.DataFrame(
            index = ['src-extract', 'build-model', 'store-model'],
            data ={
                'progress': 0,
                'err_flag': 0,
                'err_desc': None
        });

        entity:str = self.ENTITY_CUST_PROFILE;
        entity_blob:storage.Blob = ( storage.Client(project=self.GCP_PROJECT_ID)
                .bucket(self.GCP_STORAGE_BUCKET_STORAGE)
                .blob(self.ENTITY_CUST_PROFILE+'.csv') );
        src, metrix = extraction_src(entity_blob, metrix);
        metrix = ( metrix.pipe(build, src)
                .pipe(store) );

        return metrix.loc[:];


    def load_model(self)->pd.Series:
        metrix:pd.Series = pd.Series({
                    'progress':0,
                    'err_flag':0,
                    'err_desc':None
                },name='load-model');

        obj_name:str = self.ML_MODEL_CUST_CLASS+'.joblib';
        ml_blob:storage.Blob = ( storage.Client(project=self.GCP_PROJECT_ID)
                .bucket(self.GCP_STORAGE_BUCKET_ML_MODEL)
                .blob(obj_name) );

        if not ml_blob.exists():
            metrix['err_flag'] = 1;
            metrix['err_desc'] = 'Model-blob is not existed';
            return metrix.loc[:];
        try:
            ml_blob.download_to_filename(obj_name);
            self.classification = mlmodel.CustomerClassification().import_model(obj_name);
            metrix['progress'] = 1;
        except Exception as e:
            metrix['err_flag'] = 1;
            metrix['err_desc'] = str(e);
        return metrix.loc[:];


    def execute(self, customer_profile:pd.DataFrame)->pd.DataFrame:
        return self.classification.execute_model(customer_profile);

# cc = MLCustomerClassificationControl()
# cc.load_model()
# cc.execute(customer_profile=pd.read_csv('mockup_ecomm_cust_profile-202303.csv'))
```

##Pipeline Execution

```python
class PipelineExecution(PipelineControl):

    # Migration mode
    MIGRATE_MODE__SKIP:str = 'skip'; # Skip process
    MIGRATE_MODE__FULL_DUMP:str = 'full-dump'; # Full-load dataset
    MIGRATE_MODE__STRUCT_DUMP:str = 'structure-dump'; # Detected structural change - columns change
    MIGRATE_MODE__CDC_DUMP:str = 'cdc-dump'; # Detected data change
    MIGRATE_MODE__ERROR:str = 'error'; # Detected error in process

    # DW replication mode
    REPLICATE__SKIP:str = 'skip'; # Skip process
    REPLICATE__LOADED:str = 'loaded'; # Load-data success
    REPLICATE__REVERT:str = 'reverted'; # Load-data failed, revert to backup

    # Ingestion Source Files
    ingestion_src:pd.DataFrame = None;

    # Data Warehouse (BigQuery): entity-dataset mapping
    entity_bigqry_dataset_map:dict = None;

    def __init__(self):
        df:pd.DataFrame = pd.DataFrame([ self.ENTITY_PRODX_CAT, self.ENTITY_VENDOR, self.ENTITY_PRODX, self.ENTITY_CUSTOMER,
                self.ENTITY_PROMO, self.ENTITY_PRODX_PROMO, self.ENTITY_PROMO_REDEEM,
                self.ENTITY_TRX_HEADER, self.ENTITY_TRX_ITEM, self.ENTITY_TRX_PROMO], columns=['label']);
        df['local_csv'] = None;
        self.ingestion_src = df.loc[:];

        self.entity_bigqry_dataset_map = {
            # Master - Reference dataset
            self.ENTITY_PRODX_CAT: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_VENDOR: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_PRODX: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_CUSTOMER: self.GCP_BIGQRY_DATASET_REF_MASTER,

            self.ENTITY_PROMO: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_PRODX_PROMO: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_PROMO_REDEEM: self.GCP_BIGQRY_DATASET_REF_MASTER,

            self.ENTITY_TRX_HEADER: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_TRX_ITEM: self.GCP_BIGQRY_DATASET_REF_MASTER,
            self.ENTITY_TRX_PROMO: self.GCP_BIGQRY_DATASET_REF_MASTER,

            # Data warehouse
            self.ENTITY_SALE_PROD_FACT: self.GCP_BIGQRY_DATASET_FACT_DIM,
            self.ENTITY_SALE_PROMO_FACT: self.GCP_BIGQRY_DATASET_FACT_DIM,

            # Data mart
            self.ENTITY_CUST_PROFILE: self.GCP_BIGQRY_DATASET_DATA_MART,
            self.ENTITY_VEND_PROFILE: self.GCP_BIGQRY_DATASET_DATA_MART,
            self.ENTITY_ECON_METRIX: self.GCP_BIGQRY_DATASET_DATA_MART,

            # ML Result
            self.ML_RESULT_CUST_CLASS: self.GCP_BIGQRY_DATASET_DATA_MART
        }
        pass;


    def set_src(self, label:str, csv:str):
        self.ingestion_src.set_index('label', inplace=True);
        self.ingestion_src.at[label, 'local_csv'] = csv;
        self.ingestion_src.reset_index(names=['label'], inplace=True);
        return self;

    def set_src_prodx_cat(self, csv_fpath:str): return self.set_src(self.ENTITY_PRODX_CAT, csv_fpath);
    def set_src_vendor(self, csv_fpath:str): return self.set_src(self.ENTITY_VENDOR, csv_fpath);
    def set_src_prodx(self, csv_fpath:str): return self.set_src(self.ENTITY_PRODX, csv_fpath);
    def set_src_customer(self, csv_fpath:str): return self.set_src(self.ENTITY_CUSTOMER, csv_fpath);

    def set_src_promo(self, csv_fpath:str): return self.set_src(self.ENTITY_PROMO, csv_fpath);
    def set_src_prodx_promo(self, csv_fpath:str): return self.set_src(self.ENTITY_PRODX_PROMO, csv_fpath);
    def set_src_promo_redeem(self, csv_fpath:str): return self.set_src(self.ENTITY_PROMO_REDEEM, csv_fpath);

    def set_src_trx_header(self, csv_fpath:str): return self.set_src(self.ENTITY_TRX_HEADER, csv_fpath);
    def set_src_trx_item(self, csv_fpath:str): return self.set_src(self.ENTITY_TRX_ITEM, csv_fpath);
    def set_src_trx_promo(self, csv_fpath:str): return self.set_src(self.ENTITY_TRX_PROMO, csv_fpath);


    def ingest_to_rawzone(self) -> pd.DataFrame:
        df:pd.DataFrame = self.ingestion_src.loc[:];
        df['err_flag'] = 0;
        df['err_desc'] = None;
        df['timestamp'] = None;

        bucket = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_RAW);

        df.set_index('label', inplace=True);
        for i in df.index:
            fpath = df.at[i, 'local_csv'];
            bname = i+'.csv';
            try:
                blob = bucket.blob(bname);
                if not blob.exists():
                    blob.upload_from_filename(fpath);
            except Exception as e:
                df.at[i, 'err_flag']=1;
                df.at[i, 'err_desc']=e;
            df.at[i, 'timestamp']=pd.Timestamp.now();
        df.reset_index(names=['label'], inplace=True);
        return df.loc[:];


    def migrate_to_storage(self, force_update_cols:bool=False, execute=True) -> pd.DataFrame:
        def retrieve_data(df:pd.DataFrame, col_blob:str, col_df) -> pd.DataFrame:
            df = df.loc[:];
            try:
                with df[col_blob].open() as csv: df[col_df]=pd.read_csv(csv);
            except Exception as e:
                emsg:str = str(e);
                df['migrate_mode'] = self.MIGRATE_MODE__ERROR;
                df['reason'] = emsg if df['reason'] is None else '||'.join([df['reason'], emsg]);
            return df.loc[:];

        def retrieve_raw_data(df:pd.DataFrame) -> pd.DataFrame:
            return retrieve_data(df, 'blob_raw','df_raw') if (df['migrate_mode'] is None) or (df['migrate_mode'] == self.MIGRATE_MODE__FULL_DUMP) else df.loc[:];

        def retrieve_stored_data(df:pd.DataFrame) -> pd.DataFrame:
            return retrieve_data(df, 'blob_stored','df_stored') if df['migrate_mode'] is None else df.loc[:];

        def col_check(df:pd.DataFrame, force_update:bool) -> pd.DataFrame:
            if df['migrate_mode'] is not None: return df.loc[:];

            df = df.loc[:];
            rds:pd.DataFrame = df['df_raw'];
            sds:pd.DataFrame = df['df_stored'];
            rcols:pd.DataFrame = pd.DataFrame(data=rds.columns, columns=['cols']);
            rcols['r'] = 1;
            scols:pd.DataFrame = pd.DataFrame(data=sds.columns, columns=['cols']);
            scols['s'] = 1;

            cols = scols.set_index('cols').join(rcols.set_index('cols'), how='outer');
            cols['check'] = cols['r']*cols['s'];

            if len(cols[cols['check'].isna()].index) > 0:
                if not force_update:
                    df['migrate_mode'] = self.MIGRATE_MODE__ERROR;
                    df['reason'] = 'Mismatched column(s)';
                else:
                    df['migrate_mode'] = self.MIGRATE_MODE__STRUCT_DUMP;
            return df.loc[:];

        def data_check(df:pd.DataFrame):
            if df['migrate_mode'] is not None: return df.loc[:];

            df = df.loc[:];
            rds = df['df_raw'];
            sds = df['df_stored'];
            rds = rds[sds.columns].loc[:];

            r_hash:pd.DataFrame = pd.util.hash_pandas_object(rds.set_index('id'), index=False).reset_index(name='hash');
            r_hash['hash'] = r_hash['hash'].astype(str);

            s_hash:pd.DataFrame = pd.util.hash_pandas_object(sds.set_index('id'), index=False).reset_index(name='hash');
            s_hash['hash'] = s_hash['hash'].astype(str);

            hash = s_hash.set_index('id').join(r_hash.set_index('id'), how='outer', lsuffix='_stored', rsuffix='_raw').reset_index(names=['id']);
            hash['diff'] = hash['hash_stored'].ne(hash['hash_raw']);
            hash['flag'] = hash.apply(lambda x: 'unchanged' if not x['diff'] else
                    'insert' if pd.isnull(x['hash_stored']) else
                    'delete' if pd.isnull(x['hash_raw']) else 'update', axis='columns');

            if hash['diff'].sum() > 0:
                df['migrate_mode'] = self.MIGRATE_MODE__CDC_DUMP;
                df['cdc'] = hash.drop(columns=['hash_stored', 'hash_raw', 'diff']).loc[:];
            else:
                df['migrate_mode'] = self.MIGRATE_MODE__SKIP;
                df['reason'] = 'No change detected';
            return df.loc[:];

        def migrate_blob(df:pd.DataFrame, bucket_src:storage.Bucket, bucket_dst:storage.Bucket) -> pd.DataFrame:
            mode:str = df['migrate_mode'];
            try:
                if (mode == self.MIGRATE_MODE__STRUCT_DUMP) or (mode == self.MIGRATE_MODE__CDC_DUMP):
                    df['blob_stored'].delete();
                if (mode != self.MIGRATE_MODE__SKIP) and (mode != self.MIGRATE_MODE__ERROR):
                    bucket_src.copy_blob(df['blob_raw'], bucket_dst);
            except Exception as e:
                df['migrate_mode'] = self.MIGRATE_MODE__ERROR;
                df['reason'] = str(e);
                return df.loc[:];

            try:
                if df['blob_raw_exist'] and (mode != self.MIGRATE_MODE__ERROR):
                    df['blob_raw'].delete();
            except Exception as e:
                df['migrate_mode'] = self.MIGRATE_MODE__ERROR;
                df['reason'] = str(e);
            return df.loc[:];

        def report_unchanged(df:pd.DataFrame) -> int:
            mode:str = df['migrate_mode'];
            cdc:pd.DataFrame = df['cdc'];
            if (mode == self.MIGRATE_MODE__CDC_DUMP): return len(cdc[cdc['flag']=='unchanged'].index);
            return None;

        def report_insert(df:pd.DataFrame) -> int:
            mode:str = df['migrate_mode'];
            raw:pd.DataFrame = df['df_raw'];
            cdc:pd.DataFrame = df['cdc'];
            if (mode == self.MIGRATE_MODE__FULL_DUMP) or (mode == self.MIGRATE_MODE__STRUCT_DUMP): return len(raw.index);
            if (mode == self.MIGRATE_MODE__CDC_DUMP): return len(cdc[cdc['flag']=='insert'].index);
            return None;

        def report_update(df:pd.DataFrame) -> int:
            mode:str = df['migrate_mode'];
            cdc:pd.DataFrame = df['cdc'];
            if (mode == self.MIGRATE_MODE__CDC_DUMP): return len(cdc[cdc['flag']=='update'].index);
            return None;

        def report_delete(df:pd.DataFrame) -> int:
            mode:str = df['migrate_mode'];
            stored:pd.DataFrame = df['df_stored'];
            cdc:pd.DataFrame = df['cdc'];
            if (mode == self.MIGRATE_MODE__STRUCT_DUMP): return len(stored.index);
            if (mode == self.MIGRATE_MODE__CDC_DUMP): return len(cdc[cdc['flag']=='delete'].index);
            return None;

        raw = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_RAW);
        stored = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_STORAGE);
        # cdc = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_CDC);

        metrix:pd.DataFrame = self.ingestion_src[['label']].loc[:];
        metrix['migrate_mode'] = None;
        metrix['reason'] = None;
        metrix['unchanged_row'] = None;
        metrix['insert_row'] = None;
        metrix['update_row'] = None;
        metrix['delete_row'] = None;

        # Check blob pulse
        metrix['blob_raw'] = metrix['label'].apply(lambda x: raw.blob(x+'.csv'));
        metrix['blob_raw_exist'] = metrix['blob_raw'].apply(lambda x: x.exists());
        metrix['blob_stored'] = metrix['label'].apply(lambda x: stored.blob(x+'.csv'));
        metrix['blob_stored_exist'] = metrix['blob_stored'].apply(lambda x: x.exists());

        # Flag skip-processing
        metrix['migrate_mode'] = metrix['blob_raw_exist'].apply(lambda x: None if x else self.MIGRATE_MODE__SKIP);
        metrix['reason'] = metrix['migrate_mode'].apply(lambda x: 'Source is not found' if x==self.MIGRATE_MODE__SKIP else None);

        # Flag Full-load processing
        metrix['migrate_mode'] = metrix[['blob_raw_exist', 'blob_stored_exist', 'migrate_mode']].apply(lambda x:
                self.MIGRATE_MODE__FULL_DUMP if x['blob_raw_exist'] and (not x['blob_stored_exist']) else x['migrate_mode'], axis='columns');

        # Retrieve data as dataframe
        metrix['df_raw'] = None;
        metrix['df_stored'] = None;
        metrix = metrix.apply(retrieve_raw_data, axis='columns');
        metrix = metrix.apply(retrieve_stored_data, axis='columns');

        # Detected data structure (and flag full-load processing if required)
        metrix = metrix.apply(lambda x: col_check(x, force_update_cols), axis='columns');

        # Check data change and flag for loading
        metrix['cdc'] = None;
        metrix = metrix.apply(data_check, axis='columns');

        # Apply data migration
        if execute: metrix = metrix.apply(lambda x: migrate_blob(x, raw, stored), axis='columns');

        # Report
        metrix['unchanged_row'] = metrix.apply(report_unchanged, axis='columns');
        metrix['insert_row'] = metrix.apply(report_insert, axis='columns');
        metrix['update_row'] = metrix.apply(report_update, axis='columns');
        metrix['delete_row'] = metrix.apply(report_delete, axis='columns');
        metrix.drop(columns=['blob_raw', 'blob_raw_exist', 'blob_stored','blob_stored_exist', 'df_raw', 'df_stored', 'cdc'], inplace=True);

        return metrix.loc[:];


    def transform(self) -> pd.DataFrame:
        se:pd.Series = pd.Series({
                self.ENTITY_SALE_PROD_FACT: SaleProdFactTransformationControl().execute(),
                self.ENTITY_SALE_PROMO_FACT: SalePromoFactTransformationControl().execute(),
                self.ENTITY_VEND_PROFILE: VendorProfileTransformationControl().execute(),
                self.ENTITY_CUST_PROFILE: CustomerProfileTransformationControl().execute(),
                self.ENTITY_ECON_METRIX: EconMetrixTransformationControl().execute()
            }, name='metrice');

        for x in se.index: se[x]['tgt-entity'] = x;
        return pd.concat([se[x] for x in se.index]).set_index(['tgt-entity', 'metrice']);



    def upload_to_warehouse(self, entity_ls:list) -> pd.DataFrame:

        def upload(mtx:pd.DataFrame, client:bigquery.Client) -> pd.DataFrame:
            mtx = mtx.loc[:];
            src = None;
            if mtx['src_pulse'] == 0:
                mtx['err_flag'] = 1;
                mtx['err_desc'] = 'Data source is not found';
                return mtx.loc[:];
            try:
                with mtx['src_blob'].open() as csv:
                    config:bigquery.LoadJobConfig = bigquery.LoadJobConfig();
                    config.autodetect = True;
                    config.write_disposition = "WRITE_TRUNCATE";
                    (client.load_table_from_dataframe( pd.read_csv(csv),
                                    '.'.join([ mtx['tgt_dataset'], mtx['entity'] ]),
                                    job_config=config)
                            .result());
                    mtx['uploaded'] = 1;
            except Exception as e:
                mtx['err_flag'] = 1;
                mtx['err_desc'] = str(e);

            return mtx.loc[:];

        metrix:pd.DataFrame = pd.DataFrame({
            'entity': entity_ls,
            'src_pulse': [0]*len(entity_ls),
            'tgt_dataset': None,
            'uploaded': [0]*len(entity_ls),
            'err_flag': 0,
            'err_desc': None
        });

        stored = storage.Client(project=self.GCP_PROJECT_ID).bucket(self.GCP_STORAGE_BUCKET_STORAGE);
        client = bigquery.Client(project=self.GCP_PROJECT_ID);

        metrix['src_blob'] = metrix['entity'].apply(lambda x: stored.blob(x + '.csv'));
        metrix['src_pulse'] = metrix['src_blob'].apply(lambda x: 1 if x.exists() else 0);
        metrix['tgt_dataset'] = metrix['entity'].map(self.entity_bigqry_dataset_map);
        metrix = metrix.apply(lambda x: upload(x, client), axis='columns');

        metrix.drop(columns=['src_blob'], inplace=True);
        return metrix.loc[:];


    def build_model_customer_classification(self)->pd.DataFrame: return MLCustomerClassificationControl().set_k_mean__k(4).build_model();

    def classify_customer(self)->pd.DataFrame:
        def extract_src(metrix:pd.DataFrame, entity:storage.Blob):
            if not entity.exists():
                metrix.at['src-extract', 'err_flag'] = 1;
                metrix.at['src-extract', 'err_desc'] = 'Blob is not existed';
                return None, metrix.loc[:];
            try:
                with entity.open() as csv:
                    metrix.at['src-extract', 'progress'] = 1;
                    return pd.read_csv(csv), metrix;
            except Exception as e:
                metrix.at['src-extract', 'err_flag'] = 1;
                metrix.at['src-extract', 'err_desc'] = str(e);
            return None, metrix.loc[:];

        def store_result(metrix:pd.DataFrame, customer_segment:pd.DataFrame)->pd.DataFrame:
            if metrix['err_flag'].sum() > 0: return metrix.loc[:];

            se:pd.Series = pd.Series({
                'progress': 0,
                'err_flag': 0,
                'err_desc': np.nan
            }, name='store-result')

            try:
                obj_name:str = self.ML_RESULT_CUST_CLASS+'.csv';
                ( storage.Client(project=self.GCP_PROJECT_ID)
                        .bucket(self.GCP_STORAGE_BUCKET_STORAGE)
                        .blob(obj_name)
                        .upload_from_string(customer_segment.to_csv(index=False)) );
                se['progress'] = 1;
            except Exception as e:
                se['err_flag'] = 1;
                se['err_desc'] = str(e);
            metrix = pd.concat([metrix, pd.DataFrame(se).transpose()])
            return metrix.loc[:];

        metrix:pd.DataFrame = pd.DataFrame(
            index = ['src-extract'],
            data ={
                'progress': 0,
                'err_flag': 0,
                'err_desc': np.nan
        });

        # Extract Source
        src:storage.Blob = ( storage.Client(project=self.GCP_PROJECT_ID)
                        .bucket(self.GCP_STORAGE_BUCKET_STORAGE)
                        .blob(self.ENTITY_CUST_PROFILE+'.csv') )
        df, metrix = extract_src(metrix, src);
        if metrix['err_flag'].sum() > 0: return metrix.loc[:];
        # df = pd.read_csv('mockup_ecomm_cust_profile-202303.csv')

        ctrl:MLCustomerClassificationControl = MLCustomerClassificationControl();

        # Load Model
        load_status:pd.Series = ctrl.load_model();
        metrix = pd.concat( [metrix, pd.DataFrame(load_status).transpose()] );
        if metrix['err_flag'].sum() > 0: return metrix.loc[:];

        # Run model
        run_status = pd.Series({
            'progress': 0,
            'err_flag': 0,
            'err_desc': np.nan
        }, name='run');
        try:
            customer_segment:pd.DataFrame = ctrl.execute(df);
            run_status['progress']=1;
        except Exception as e:
            run_status['err_flag']=1;
            run_status['err_desc']=str(e);
        metrix = pd.concat( [metrix, pd.DataFrame(run_status).transpose()] );
        if metrix['err_flag'].sum() > 0: return metrix.loc[:];

        metrix = store_result(metrix, customer_segment);
        return metrix.loc[:]
```
